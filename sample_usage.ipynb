```python
import time
import matplotlib.pyplot as plt
from kaolin_render import render_glb_with_hammersley
```

```python
glb_path = "PATH/TO/YOUR/MODEL.glb"

start = time.time()
images, depths = render_glb_with_hammersley(glb_path, num_views=10, res=512)
end = time.time()
print(f"Rendering took {end - start:.2f} seconds")
```

```python
plt.figure(figsize=(15, 5))
for i, img in enumerate(images):
    plt.subplot(2, 5, i + 1)
    plt.imshow(img)
    plt.axis('off')
```

```python
plt.figure(figsize=(15, 5))
for i, img in enumerate(depths):
    plt.subplot(2, 5, i + 1)
    plt.imshow(img.squeeze(), cmap='gray')
    plt.axis('off')
```
