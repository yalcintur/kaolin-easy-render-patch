```python
import time
import matplotlib.pyplot as plt
from kaolin_render import render_glb_with_hammersley
```

```python
glb_path = "/home/ubuntu/lacd/acid/data/train_data/behavior1k/behavior1k_glbs/apple__obixxh.glb"

images, depths = render_glb_with_hammersley(glb_path, num_views=10, res=512)
```
